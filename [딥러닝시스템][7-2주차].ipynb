```python
import pandas as pd
```

```python
train_x = pd.read_csv("/kaggle/input/dataset/train_x.csv")
train_y = pd.read_csv("/kaggle/input/dataset/train_y.csv")
test_x = pd.read_csv("/kaggle/input/dataset/test_x.csv")
test_y = pd.read_csv("/kaggle/input/dataset/test_y.csv")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,LSTM,Embedding
from tensorflow.keras.utils import to_categorical
```

```python
tra_x = list()
tra_y = list()
tes_x = list()
tes_y = list()
```

```python
tra_y = train_y.iloc[:,1]
tes_y = test_y.iloc[:,1]
```

```python
import numpy as np
```

```python
from ast import literal_eval
for i in range(len(train_x)):
    tra_x.append(np.array(literal_eval(train_x.iloc[i][1])))
    
for i in range(len(test_x)):
    tes_x.append(np.array(literal_eval(test_x.iloc[i][1])))
```

```python
import numpy as np
# tra_x = np.array(tra_x)
tra_y = np.array(tra_y)
# tes_x = np.array(tes_x)
tes_y = np.array(tes_y)
```

```python
word_size = 1000
```

```python
# model = Sequential()
# model.add(Embedding(word_size, 4, input_length=3))
# model.add(LSTM(3))
# model.add(Flatten())
# model.add(Dense(1, activation='sigmoid'))
# model.summary()

model = Sequential()
model.add(Embedding(word_size, 4))
model.add(LSTM(3))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
from keras.preprocessing.sequence import pad_sequences

# Padding을 사용하여 시퀀스의 길이를 맞춤
max_length = max(len(seq) for seq in tra_x)
tra_x = pad_sequences(tra_x, maxlen=max_length, padding='post')

print("Padded tra_x shape:", tra_x.shape)
```

```python
maxt_length = max(len(seq) for seq in tes_x)
tes_x = pad_sequences(tes_x, maxlen=maxt_length, padding='post')

print("Padded tra_x shape:", tes_x.shape)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(tra_x, tra_y, epochs=5)
```

```python
print("\n Accuracy: %.4f" % (model.evaluate(tes_x, tes_y)[1]))
```
